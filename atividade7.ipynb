```python
import whisper
```

```python
model = whisper.load_model("large")
```

```python
result = model.transcribe("/content/videoplayback.m4a")
print(result['text'])
```

```python
from transformers import pipeline
```

```python
summarizer = pipeline("summarization", model="google/flan-t5-xl") # you can replace "google/flan-t5-xl" with any other llma 2 model
```

```python
summary = summarizer(result['text'], max_length=130, min_length=30, do_sample=False)
print(summary[0]["summary_text"])
```
